```python
import pandas as pd
from dython.nominal import associations

df = pd.read_parquet('data/wine/wine_review.parquet.gzip', columns=['winery', 'year', 'variety', 'points', 'taster_name', 'country', 'location', 'type', 'wine_style',	'quality',	'classification'])
print(df.shape)
df.head()
```

```python
from dython.nominal import identify_nominal_columns
categorical_features=identify_nominal_columns(df)
categorical_features
```

```python
complete_correlation= associations(df, filename= 'complete_correlation.png', figsize=(10,10))
```

```python
df_complete_corr=complete_correlation['corr']
df_complete_corr.dropna(axis=1, how='all').dropna(axis=0, how='all').style.background_gradient(cmap='coolwarm', axis=None).format('{:.2f}')
```
